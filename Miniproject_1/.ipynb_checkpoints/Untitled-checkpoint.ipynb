The goal is to find a model $\phi$:
$$\phi(\hat{X}) = X$$
Quadratic loss: 
$$\hat{w} = argmin \frac{1}{N} \sum_{n=1}^N || x_n - \phi(x_n + \epsilon_n;w) || ^2$$
* $x_n$: data samples
* $\epsilon_n$ : gaussian noise

```python
from torch import nn
from tqdm import tqdm
model = nn.Sequential(
    nn.Linear(32, 100),
    nn.ReLU(),
    nn.Linear(100,32)
)

batch_size, nb_epochs = 100, 10
optimizer = torch.optim.Adam(model.parameters(), lr = 1e-3)
mse = nn.MSELoss()

data = noisy_imgs_1
for e in tqdm(range(nb_epochs)):
    input = noisy_imgs.type(torch.float).split(batch_size)
    target = clean_imgs.type(torch.float).split(batch_size)
    for i in range(len(input)):
        output = model(input[i])
        
        loss = mse(output[i], target[i])
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
#         plt.figure()
#         plt.subplot(1, 2, 1)
#         plt.imshow(input[0][0, 0])
#         plt.subplot(1, 2, 2)
#         plt.imshow(target[0][0, 0])
        
#     plt.figure()
#     plt.subplot(1, 3, 1)
#     plt.imshow(input[0][0, 0])
#     plt.subplot(1, 3, 2)

#     plt.imshow(output[0, 0].detach());
#     plt.subplot(1, 3, 3)
#     plt.imshow(target[0][0,0])
    
        
```
